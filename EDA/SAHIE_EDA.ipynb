```python
sahie['state_name'] = sahie['state_name'].str.strip()
sahie['county_name'] = sahie['county_name'].str.strip()
sahie[
    (sahie['state_name'] == 'California') &
    (sahie['county_name'] == 'Orange County')
    ]
```

```python
# kalawao county is documented to have invalid/null data
sahie[
    (sahie['county_name'] == 'Kalawao County')
    ]
```

```python
sahie = sahie[
    (sahie['county_name'] != 'Kalawao County')
    ]
```
