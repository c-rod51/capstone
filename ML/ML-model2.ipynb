# Further Tuning of the ML Model for the InsuranceCharges Data

Since we saw that the model does not perform differently including sex, we will remove it from our data. In `ML-model.ipynb` we saw that the LinearRegression (i.e. Lasso with $\alpha$ = 0) was the best performing model.

In this notebook we will try to improve the accuracy of the LinearRegression model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso, LassoCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('clean_insurance.csv')
```

```python
df = df.drop(columns = 'sex')
```

```python
df
```

```python
df_dummies = pd.get_dummies(df, columns = ['region'], drop_first = True)
```

```python
X = df_dummies.drop(columns = 'charges').copy()
y = df_dummies[['charges']].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 9)
```

```python
model = make_pipeline(PolynomialFeatures(degree=2),
                      LinearRegression())
```

```python
param_grid = {'polynomialfeatures__degree': np.arange(8),
              'linearregression__fit_intercept': [True, False],
              'linearregression__normalize': [True, False]}

grid = GridSearchCV(model, param_grid, cv=7)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

The best params are
* fit_intercept: True
* normalize: True
* polynomialfeatures_degree: 2

```python
best_model = grid.best_estimator_
```

```python
best_model.score(X_test, y_test)
```

With approximately 9 percentage point increase the new model has an accuracy of 86.6%

This was mostly achieved through the use of polynomialfeatures as we saw the alphas made little-to-no difference near $\alpha$ = 0

We can use the model to create predictions for potential insurance customers, but first we export the model using joblib
